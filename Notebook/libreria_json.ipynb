# JSON
Sono un formato che viene utilizzato molto per comunicare con i server web e per le configurazioni dei programmi.

Sta **J**ava**S**cript **O**bject **N**otation ed è una struttura di dati chiave valore.

La libreria ci aiuta a interpretare in maniera corretta i file json per estrarre informazioni e utilizzarle nel nostro script

## Json dumps

Permette di prendere un oggetto di python e lo trasforma in un JSON sotto forma di stringa.



```python
test = {"Prova": "1", "config": ["q", "w", "e", "r", "t", "y"]}
print("Oggetto test di tipo", type(test))
test_dumps = json.dumps(test)
print("test_dumps:", test_dumps, "\ndi tipo", type(test_dumps))
```

```python
test_pretty = json.dumps(obj=test, indent=4)  # , separators=("°", "$"))
print(test_pretty)
```

```python
test_list = [1,2,3,4,5,6]
json.dumps(test_list)
```

## Json loads

Genera un oggetto python da una stringa JSON


```python
j_str = '{"Prova": "1", "config": ["q", "w", "e", "r", "t", "y"]}'
j_dict = json.loads(j_str)
j_dict.items()
```

## Lavorare con i File

I metodi che abbiamo visto ci permettono di passare da JSON sotto forma di stringa a oggetti di python. 

Per leggere e scrivere su file `.json`, esistono i metodi `dump` e  `load` che funzionano in maniera analoga ma richiedono un parametro aggiuntivo che è un oggetto di tipo file.

```python
## scrivere file json con dumps
with open("test_json.json", 'w') as json_file:
    json.dump(obj=test, fp=json_file, indent=4)
```
